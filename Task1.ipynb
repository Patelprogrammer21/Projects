```python
import requests
from bs4 import BeautifulSoup
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd

url = "https://www.equitymaster.com/india-markets/nse-replica.asp"
r = requests.get(url) # r variable has all the HTML code
htmlContent = r.content # r returns response so if we want the code we write r.content
# print(htmlContent)

htmlText = r.text
# print(htmlText)

soup = BeautifulSoup(r.content, 'html.parser')

# Assuming `soup` is your BeautifulSoup object containing the HTML
table_data = []
col_name = []

# Loop through each table with the specified class
for i in soup.find_all("table", class_="mystocksbig mystocks newtab cls"):
    rows = i.find_all('tr')  # Find all rows in the table
    for row in rows:
        cols = row.find_all('th')  # Find all columns (th elements)
        cols = [col.text.strip() for col in cols]  # Extract text and remove extra spaces
        col_name.append(cols)
        cols = row.find_all('td')  # Find all columns (td elements)
        cols = [col.text.strip() for col in cols]  # Extract text and remove extra spaces
        table_data.append(cols)  # Append the row data to the table_data list

col_name = col_name[1:3]
col_name = sum(col_name, []) # converting list of list to a single list
del col_name[6]

# Convert the list of lists into a pandas DataFrame
df = pd.DataFrame(data=table_data, columns=col_name)
df = df.dropna(subset="COMPANY").reset_index(drop = True)

print(df.head())
# df.to_csv('/content/drive/My Drive/AbhiData/output.csv', index=False)
```
