```python
import sqlite3
import pandas as pd

# Step 1: Read the CSV file into a Pandas DataFrame
csv_file_path = '/content/drive/My Drive/AbhiData/HEROMOTOCO.NS.csv'  # Update with your CSV file path
df = pd.read_csv(csv_file_path)

# Step 2: Connect to (or create) an SQLite database
db_path = 'csv_to_sqlite.db'  # SQLite DB name
conn = sqlite3.connect(db_path)
cursor = conn.cursor()

# Step 3: Create a table dynamically based on CSV columns
table_name = 'my_table'  # You can choose your table name
columns = ', '.join([f'{col} TEXT' for col in df.columns])  # Assuming all columns are text type
create_table_query = f'CREATE TABLE IF NOT EXISTS {table_name} ({columns});'

cursor.execute(create_table_query)

# Step 4: Insert data into the table
for row in df.itertuples(index=False):
    placeholders = ', '.join(['?' for _ in df.columns])  # Create placeholders for column values
    insert_query = f'INSERT INTO {table_name} VALUES ({placeholders});'
    cursor.execute(insert_query, tuple(row))  # Insert each row

# Step 5: Commit changes and close the connection
conn.commit()
# conn.close()

print(f"Database '{db_path}' created with table '{table_name}' from the CSV file.")
```
