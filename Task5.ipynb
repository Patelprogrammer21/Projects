```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

headers = {
    'User-Agent': 'Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/128.0.0.0 Mobile Safari/537.36'
}

url = 'https://www.forbes.com/lists/global2000/'
response = requests.get(url, headers=headers)

details = []
soup = BeautifulSoup(response.content, 'html.parser')

for i in soup.find_all("div", class_="table-row-group"):
    rows = i.find_all('a')
    for row in rows:
        cols = row.find_all('div')
        cols = [col.text.strip() for col in cols]
        link = row.get('href')
        company_name = cols[1]
        if(link is None):
          continue
        else:
          ceo_name = function(link)
          record = [company_name, ceo_name]
          details.append(record)
          # print(record)
df = pd.DataFrame(details)
# df.head()
df.to_csv(f'/content/drive/My Drive/AbhiData/company_with_ceo.csv', index=False)
```
